```python
def mf1(profit_value, quantile25, quantile75):
    if profit_value<quantile25:
        return "avg profit"
    elif profit_value>quantile75:
        return "excellent profit"
    else:
        return "good profit"
```

```python
df1['Profit']=(df1.Rev-df1.Exp)
df1
```

```python
q25=(df1.Rev-df1.Exp).quantile(0.25)
q75=(df1.Rev-df1.Exp).quantile(0.75)
print(q25,q75)
df1['Profit_class']=(df1.Rev-df1.Exp).apply(func=mf1,args=(q25,q75))
df1
```
