```python
df=sns.load_dataset('tips')
df.head()
```

```python
sns.relplot(x="total_bill",y="tip",data=df)#data=df is imp as it helps us to give the reference to which it belongs
plt.show()
#SCATTER PLOT HELPS US TO BUILD SOME KIND OF A RELATIONSHIP
#AS TOTAL BILL INCREASES TOTAL TIP INCREASES
```

```python
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter")#data=df is imp as it helps us to give the reference to which it belongs
plt.show()
#SCATTER PLOT HELPS US TO BUILD SOME KIND OF A RELATIONSHIP
#AS TOTAL BILL INCREASES TOTAL TIP INCREASES
```

```python
sns.relplot(x="total_bill",y="tip",data=df,kind="line")#data=df is imp as it helps us to give the reference to which it belongs
plt.show()
#SCATTER PLOT HELPS US TO BUILD SOME KIND OF A RELATIONSHIP
#AS TOTAL BILL INCREASES TOTAL TIP INCREASES
```

# CATEGORIAL SCATTER PLOT

#CATEGORIAL RELATIONSHIP

1.catplot() fn is used to analysze the relationship between a numeric value 
and a categorial group of values together

SYNTAX

seaborn.catplot(x=value,y=value,data=data)
2. categorial data is represented in x axis and values correspond to them reprsented through y-axis(standard notation).we can reverse it.
3. the defualt representation of the data in catplot() uses a scatterplot(stripplot)

```python
df=sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(x='sex',y='total_bill',data=df) # X IS CATEGORIAL Y IS NUMERICAL
plt.show()
```

STRIPPLOT()
1.a strip plot is a scatter plot where one of the variables is categorial
2.and for each category in the categorial variable,we see a scatter plot with respect to the numeric column

PARAMETERS-
1.the first parameter is the categorial column,the second column is the numeric column while the third is dataset.
2. the jitter parameter controls the magnitude of jitter or disables it altogether.jitter is the deviation from true value


```python
sns.catplot(x='sex',y='total_bill',data=df,jitter=0.5) # X IS CATEGORIAL Y IS NUMERICAL
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,jitter=0.25) # X IS CATEGORIAL Y IS NUMERICAL
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,jitter=0.005) # X IS CATEGORIAL Y IS NUMERICAL
plt.show()
#HERE THERE IS LESS RANDOMNESS FROM THE TRUE VALUE
```

# SWARMPLOT()

similar to stripplot,only difference is that it does not allow overlapping of markers.
2.it adjusts the points along the categorial axis using an algo that prevents them from overlapping.
3.it can give a better representation of the distribution of observations.
4.although it only works well for relatively small datasets.because sometimes it takes a lot of computation to arrange large datasets such that they are not overlapping.


```python
sns.catplot(x='sex',y='total_bill',data=df,kind="swarm") # X IS CATEGORIAL Y IS NUMERICAL
plt.show()
#HERE THERE IS LESS RANDOMNESS FROM THE TRUE VALUE
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind="swarm",hue='smoker') # X IS CATEGORIAL Y IS NUMERICAL
plt.show()
#HERE THERE IS LESS RANDOMNESS FROM THE TRUE VALUE
```

# BOXEN PLOT

SIMILAR TO BOX PLOT IT PLOT DIFFERENT QUARTIKLE VALUES

# VIOLIN PLOT

#VIOLIN PLOT BAUSCALLY GIVES KDE OF THE DISTRIBUTION.
1.IT SHOWS THE DISTRIBUTION OF QUANTATIVE DATA ACROSS SEVERAL LEVELS OF ONE (OR MORE) CATEGORIAL VARIABLES SUCH THAT THOSE DISTRBUTION CAN BE COMPARED.
2. IT IS REALLY CLOSE FROM A BOXPLOT,BUT ALLOWS A DEEPER UNDERSTANDING OF THE DENSITY
3. UNLIKE A BOX PLOT,IN WHICH ALL OF THE PLOT COMPONENTS COREESPONDS TO ACTUAL DATA POINTS,THE VIOLIN PLOT FEATURES A KERNEL DENSITY ESTIMATION OF THE UNDERLYING DISTRIBUTION.
4.  THE KDE IS A MATHEMATICAL PROCESS OF FINDING AN ESTIMATE PROBABILITY DENSITY FN OF A RANDOM VARIABLE. 



```python
df=sns.load_dataset('iris')
df.head()
```

```python
sns.catplot(x="species",y="sepal_length",data=df,kind="violin")
plt.show()
#in sentosa more values are distributed towards the value 5 hence waha pe kde zyada h
```

THE DENSITY IS MIRRORED AND FLIPPED OVER AND THE RESULTING SHAPE IS FILLED IN,CREATING AN IMAGE RESEMBLING A VIOLIN.

BOXPLOT CLEARLITY SHOWS OUTLIERS
VIOILIN ARE PARTICULARY ADAPTED WHEN THE AMT OF DATA IS HUGE AND SHOWING INDIVIDUAL OBSERVATINS GETS IMPOSSIBLE.
